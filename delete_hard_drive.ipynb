```python
import os
import time

# Specify the folder containing the files
hard_drive_path = 'D:\\'
output_train_path = ['ignore', 'train']
output_train_folder_waves = os.path.join(hard_drive_path, *output_train_path, 'waves')
folder_path = output_train_folder_waves
batch_size = 100  # Adjust batch size as needed
pause_time = 2  # Pause in seconds between batches

def delete_files_in_batches(folder_path, batch_size, pause_time):
    # Get all files in the specified folder
    files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]

    # Delete files in batches
    for i in range(0, len(files), batch_size):
        batch = files[i:i + batch_size]
        for file in batch:
            try:
                os.remove(file)
                print(f"Deleted: {file}")
            except Exception as e:
                print(f"Failed to delete {file}: {e}")

        print(f"Batch {i // batch_size + 1} completed. Pausing for {pause_time} seconds.")
        time.sleep(pause_time)

    print("All files deleted successfully.")

# Run the deletion function
delete_files_in_batches(folder_path, batch_size, pause_time)
```
