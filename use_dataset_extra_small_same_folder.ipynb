````python
from datasets import load_dataset
dataset_repo = "johnlohjy/imda_nsc_p3_same_closemic_extra_small"
dataset_train = load_dataset(dataset_repo, split='train', streaming=True, trust_remote_code=True)
dataset_test = load_dataset(dataset_repo, split='test', streaming=True, trust_remote_code=True)
````

````python
dataset_train_iter = iter(dataset_train)
````
